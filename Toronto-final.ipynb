### 1. Import libraries

```python
# Importing libraries
import pandas as pd
import requests
from bs4 import BeautifulSoup as bs
```

### 2. Scrape data from Wikipedia page into a dataframe

```python
# Get the table of postal codes of Canada from the wiki page
page = requests.get("https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M") # Wikipedia page                                                  
soup = bs(page.text, "lxml") # Parse data from "page" 
table = soup.find("table", class_ = "wikitable") # Pinpointing the location of the table and its contents
```

```python
## Create a dataframe

# Creating lists for each of the columns in the table.
A = []
B = []
C = []

# Utilizing HTML tags for rows <tr> and elements <td> to iterate through each row of data 
# and append data elements to their appropriate lists
for row in table.findAll("tr"):
    cells = row.findAll('td')
    if len(cells) == 3: # Only extract table body not header
        A.append(cells[0].find(text = True))
        B.append(cells[1].find(text = True))
        C.append(cells[2].find(text = True))

#Coverting to dataframe
df = pd.DataFrame(A, columns = ["Postcode"])
df["Borough"] = B
df["Neighborhood"] = C       
```

```python
# Display all rows of the dataframe
pd.set_option('display.max_rows', None)
print(df.shape)
df
```

### 3. Drop cells where 'Borough" is listed as "Not assigned."

```python
# Drop rows where Borough is not assigned
df.drop(df[df.Borough == "Not assigned"].index, inplace=True)
print(df.shape)
df
```

### 4. Group neighborhoods in the same Borough

```python
# Combining Neighborhood with the same Postcode using comma join
df.replace("\n","", regex = True, inplace = True) # get rid of \n in the dataframe
new_df = df.groupby(["Postcode", "Borough"])["Neighborhood"].apply(", ".join).reset_index()
print(new_df.shape)
new_df 
```

### 5. For neighborhood listed as "Not assigned", designate a value that matches "Borough."

```python
# Not assigned Neighborhood named as Borough
new_df.loc[new_df.Neighborhood == "Not assigned", ["Neighborhood"]] = new_df["Borough"]
print(new_df.shape)
new_df
```

### 12. Explore Toronto's neighbourhoods using Foursquare data.

```python
# define Foursquare Credentials and Version
CLIENT_ID = "Your Client ID"
CLIENT_SECRET = "Your Client Secret" 
Version: "20180605"
    
print('Your credentails:')
print('CLIENT_ID: ' + CLIENT_ID)
print('CLIENT_SECRET:' + CLIENT_SECRET)
```

```python
# A function to collect nearby (radius of 500 meters) venues in all neighboroods of Toronto 

def getNearbyVenues(names, latitudes, longitudes, radius = 500):
    
    venues_list = []
    for name, lat, lng in zip(names, latitudes, longitudes):
        print(name)
            
        # create the API request URL
        url = "https://api.foursquare.com/v2/venues/explore?&client_id={}&client_secret={}&v={}&ll={},{}&radius={}".format(
            CLIENT_ID, 
            CLIENT_SECRET, 
            VERSION, 
            lat, 
            lng, 
            radius)
            
        # make the GET request
        results = requests.get(url).json()["response"]["groups"][0]["items"]
        
        # return only relevant information for each nearby venue
        venues_list.append([(
            name, 
            lat, 
            lng, 
            v["venue"]["name"], 
            v["venue"]["location"]["lat"], 
            v["venue"]["location"]["lng"],  
            v["venue"]['categories'][0]["name"]) for v in results])

    nearby_venues = pd.DataFrame([item for venue_list in venues_list for item in venue_list])
    nearby_venues.columns = ["Neighborhood", 
                  "Neighborhood Latitude", 
                  "Neighborhood Longitude", 
                  "Venue", 
                  "Venue Latitude", 
                  "Venue Longitude", 
                  "Venue Category"]
    
    return(nearby_venues)
```

```python
# Create a new dataframe called toronto_venues
toronto_venues = getNearbyVenues(names=toronto_data["Neighborhood"],
                                   latitudes=toronto_data["Latitude"],
                                   longitudes=toronto_data["Longitude"]
                                  )
```

```python
# Check the toronto_venues dataframe 
print(toronto_venues.shape)
toronto_venues.head()
```

### 13. Show the number of venues per Toronto neighbourhood.

### 14. List the number and type of venues found in each neighbourhood.

```python
# one hot encoding
toronto_onehot = pd.get_dummies(toronto_venues[["Venue Category"]], prefix = "", prefix_sep = "")

# add neighborhood column back to dataframe
toronto_onehot["Neighborhood"] = toronto_venues["Neighborhood"] 

# move neighborhood column to the first column
fixed_columns = [toronto_onehot.columns[-1]] + list(toronto_onehot.columns[:-1])
toronto_onehot = toronto_onehot[fixed_columns]

print(toronto_onehot.shape)
toronto_onehot.head()
```

```python
# Group rows by neighborhood and by taking the mean of the frequency of occurrence of each category
toronto_grouped = toronto_onehot.groupby("Neighborhood").mean().reset_index()

print(toronto_grouped.shape)
toronto_grouped.head()
```

### 15. List venue types by neighborhood, including the top five for each.

```python
# print each neighborhood along with the top 5 most common venues

num_top_venues = 5

for hood in toronto_grouped["Neighborhood"]:
    print("----"+hood+"----")
    temp = toronto_grouped[toronto_grouped["Neighborhood"] == hood].T.reset_index()
    temp.columns = ["venue","freq"]
    temp = temp.iloc[1:]
    temp["freq"] = temp["freq"].astype(float)
    temp = temp.round({"freq": 2})
    print(temp.sort_values("freq", ascending = False).reset_index(drop = True).head(num_top_venues))
    print('\n')
```

```python
# A function to sort the venues in descending order

def return_most_common_venues(row, num_top_venues):
    row_categories = row.iloc[1:]
    row_categories_sorted = row_categories.sort_values(ascending = False)
    
    return row_categories_sorted.index.values[0:num_top_venues]
```

### 16. Create a dataframe with Neighborhoods as rows, with the columns representing the top 1-10 venue types for each (in descending order).

```python
# Create a new dataframe of neighborhoods with the top 10 venues for each neighborhood

import numpy as np

num_top_venues = 10

indicators = ['st', 'nd', 'rd']

# create columns according to number of top venues
columns = ["Neighborhood"]
for ind in np.arange(num_top_venues):
    try:
        columns.append("{}{} Most Common Venue".format(ind+1, indicators[ind]))
    except:
        columns.append("{}th Most Common Venue".format(ind+1))

# create a new dataframe
neighborhoods_venues_sorted = pd.DataFrame(columns = columns)
neighborhoods_venues_sorted["Neighborhood"] = toronto_grouped["Neighborhood"]

for ind in np.arange(toronto_grouped.shape[0]):
    neighborhoods_venues_sorted.iloc[ind, 1:] = return_most_common_venues(toronto_grouped.iloc[ind, :], num_top_venues)

neighborhoods_venues_sorted
```

### 17. Perform a k-means cluster analysis based on the mix of venues within each neighbourhood.

```python
# import k-means from clustering stage
from sklearn.cluster import KMeans

# set number of clusters
kclusters = 5

toronto_grouped_clustering = toronto_grouped.drop("Neighborhood", 1)

# run k-means clustering
kmeans = KMeans(n_clusters=kclusters, random_state=0).fit(toronto_grouped_clustering)

# check cluster labels generated for each row in the dataframe
kmeans.labels_[0:100] 
```

```python
# Create a new dataframe that includes the clusters as well as the top 10 venues for each neighborhood

toronto_merged = toronto_data

# add clustering labels
toronto_merged["Cluster Labels"] = kmeans.labels_

# merge neighborhoods_venues_sorted with tor_df to add latitude/longitude for each neighborhood
toronto_merged = toronto_merged.join(neighborhoods_venues_sorted.set_index("Neighborhood"), on = "Neighborhood")

print(toronto_merged.shape)
toronto_merged
```

### 18. Now, put the clusters on a map of Toronto.

```python
# Visualize the resulting clusters

# Matplotlib and associated plotting modules
import matplotlib.cm as cm
import matplotlib.colors as colors

# create map
map_clusters = folium.Map(location=[latitude, longitude], zoom_start = 11)

# set color scheme for the clusters
x = np.arange(kclusters)
ys = [i + x + (i*x)**2 for i in range(kclusters)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map
markers_colors = []
for lat, lon, poi, cluster in zip(toronto_merged["Latitude"], toronto_merged["Longitude"], toronto_merged["Neighborhood"], toronto_merged["Cluster Labels"]):
    label = folium.Popup(str(poi) + "Cluster" + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius = 5,
        popup = label,
        color = rainbow[cluster-1],
        fill = True,
        fill_color = rainbow[cluster-1],
        fill_opacity = 0.7).add_to(map_clusters)
       
map_clusters
```
